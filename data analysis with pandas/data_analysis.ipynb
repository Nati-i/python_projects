```python
def ret_lat(address):
    a = Nominatim()
    a = a.geocode(address)
    if a != None:
        return a.latitude
    else:
        return None

def ret_lon(address):
    a = Nominatim()
    a = a.geocode(address)
    if a != None:
        return a.longitude
    else:
        return None
```
